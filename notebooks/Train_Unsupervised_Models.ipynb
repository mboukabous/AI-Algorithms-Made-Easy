- **`--results_path`**: Path to save any results (like cluster labels, dimensionality-reduced data, outlier predictions, plots).  
- **`--visualize`**: Generate and save 2D plots (via PCA) for clusters/outliers, or a 2D/3D plot for dimension reduction.

### 2.3.1 Clustering
<a id="clustering"></a>

Clustering algorithms group data points based on their similarity, **without** requiring labeled examples. They aim to discover inherent structures or patterns in data. Clustering is commonly used for **customer segmentation**, **image segmentation**, and **grouping similar documents**.


#### K-Means
<a id="kmeans"></a>

**Description**:  
K-Means partitions the dataset into `k` clusters by assigning each data point to the cluster with the nearest centroid. It then recalculates the centroid of each cluster iteratively until convergence.

**Key Characteristics**:  
- **Simple and Fast**: Works well on moderate-size numeric data.  
- **Requires k**: Must specify number of clusters in advance.  
- **Sensitive to Scale/Outliers**: Often benefits from feature scaling.


##### Define the model (optional)

````python
%%writefile models/unsupervised/clustering/kmeans.py

"""
kmeans.py

This module defines a KMeans clustering model and a parameter grid for hyperparameter tuning.

KMeans is a popular clustering algorithm that partitions data into k clusters. Each cluster is represented by the centroid of its members, and the algorithm iteratively refines the centroids to minimize the within-cluster variance.

Parameters:
    - n_clusters (int): Number of clusters to form.
    - init (str): Initialization method for centroids. Common options:
        - 'k-means++' (default): Optimized centroid initialization.
        - 'random': Random initialization.
    - n_init (int): Number of times the algorithm runs with different centroid seeds.
    - random_state (int): Ensures reproducibility of results.
"""

from sklearn.cluster import KMeans

# Define the KMeans estimator
estimator = KMeans(n_clusters=3, random_state=42)

# Define the hyperparameter grid for tuning
param_grid = {
    'model__n_clusters': [2, 3, 4, 5],  # Experiment with 2 to 5 clusters
    'model__init': ['k-means++', 'random'],  # Compare optimized and random initialization
    'model__n_init': [10, 20, 50]  # Test different numbers of initializations for stability
}

# Use silhouette score as the default scoring metric
# Silhouette score evaluates how well clusters are separated and compact
default_scoring = 'silhouette'
````

#### DBSCAN
<a id="dbscan"></a>

**Description**:  
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) forms clusters based on dense regions of data points, labeling points in low-density regions as outliers.

**Key Characteristics**:  
- **No Predefined k**: Discovers number of clusters automatically.  
- **Robust to Outliers**: Naturally identifies noisy points.  
- **Sensitivity to Parameters**: Requires `eps` (neighborhood size) and `min_samples` to be well-chosen.


##### Define the model (optional)

````python
%%writefile models/unsupervised/clustering/dbscan.py

"""
dbscan.py

This module defines a DBSCAN clustering model and a parameter grid for hyperparameter tuning.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm.
It groups points closely packed together and marks as outliers those points in low-density regions.

Parameters:
    - eps (float): The maximum distance between two samples for them to be considered as in the same neighborhood.
    - min_samples (int): The number of samples (or total weight) in a neighborhood for a point to be considered a core point.
"""

from sklearn.cluster import DBSCAN

# Define the DBSCAN estimator
estimator = DBSCAN(eps=0.5, min_samples=5)

# Define the hyperparameter grid for tuning
param_grid = {
    'model__eps': [0.2, 0.5, 1.0, 1.5, 2.0],  # Explore a wide range of neighborhood radii
    'model__min_samples': [3, 5, 10, 20]  # Adjust density thresholds for core points
}

# Default scoring metric
# Note: Silhouette score works best for convex clusters and may not always be ideal for DBSCAN.
# For more complex shapes, consider custom evaluation metrics.
default_scoring = 'silhouette'
````

#### Gaussian Mixture
<a id="gaussian-mixture"></a>

**Description**:  
A probabilistic model that assumes data is generated from a mixture of Gaussian distributions. Each cluster corresponds to one Gaussian component.

**Key Characteristics**:  
- **Soft Assignments**: Points have probabilities of belonging to each cluster.  
- **Flexible**: Can model complex cluster shapes if you allow covariance structures.  
- **May Need k**: Usually requires specifying number of components.


##### Define the model (optional)

````python
%%writefile models/unsupervised/clustering/gaussian_mixture.py

"""
gaussian_mixture.py

This module defines a GaussianMixture model for clustering, along with a parameter grid for hyperparameter tuning.

Gaussian Mixture Models (GMM) assume that data is generated from a mixture of several Gaussian distributions
with unknown parameters. It's a probabilistic model and can handle clusters of varying sizes and shapes.

Parameters:
    - n_components (int): Number of mixture components (clusters).
    - covariance_type (str): Determines the shape of each cluster.
        - 'full': Each cluster has its own general covariance matrix.
        - 'tied': All clusters share the same covariance matrix.
        - 'diag': Each cluster has its own diagonal covariance matrix.
        - 'spherical': Each cluster has its own single variance.
"""

from sklearn.mixture import GaussianMixture

# Define the GaussianMixture estimator
estimator = GaussianMixture(n_components=3, random_state=42)

# Define the hyperparameter grid for tuning
param_grid = {
    'model__n_components': [2, 3, 4],  # Experiment with 2 to 4 clusters
    'model__covariance_type': ['full', 'tied', 'diag', 'spherical']  # Different shapes for cluster covariance
}

# Default scoring metric
# Note: Silhouette score works better for convex clusters. For GMMs with non-convex clusters, consider other metrics like BIC or AIC.
default_scoring = 'silhouette'
````

##### Execute the model

#### Hierarchical Clustering
<a id="hierarchical-clustering"></a>

**Description**:  
Hierarchical (often agglomerative) clustering merges the closest pairs of clusters step by step, forming a hierarchy (dendrogram). You can choose how many clusters to keep by “cutting” the dendrogram at a specific level.

**Key Characteristics**:
- **No Need for k Up Front**: You can pick the number of clusters afterward by cutting the dendrogram.  
- **Dendrogram**: Offers a tree-based view of cluster merge steps, aiding interpretability.  
- **Linkage Criteria**: Common linkage methods (single, complete, average, Ward) affect how distances between clusters are computed.


##### Define the model (optional)

````python
%%writefile models/unsupervised/clustering/hierarchical_clustering.py

"""
hierarchical_clustering.py

This module defines an AgglomerativeClustering model for hierarchical clustering,
along with a parameter grid for hyperparameter tuning.

Hierarchical clustering creates a tree-like structure (dendrogram) to represent the nested grouping of data points
and their similarity levels. Agglomerative clustering starts with each data point as its own cluster and iteratively merges them.

Parameters:
    - n_clusters (int): The number of clusters to form.
    - linkage (str): Determines how distances between clusters are computed.
        - 'ward': Minimizes the variance of clusters (requires Euclidean distance).
        - 'complete': Maximum linkage, i.e., uses the farthest points between clusters.
        - 'average': Average linkage, i.e., uses the mean distances between clusters.
        - 'single': Minimum linkage, i.e., uses the closest points between clusters.
"""

from sklearn.cluster import AgglomerativeClustering

# Define the AgglomerativeClustering estimator
estimator = AgglomerativeClustering(n_clusters=3)

# Define the hyperparameter grid for tuning
param_grid = {
    'model__n_clusters': [2, 3, 4],  # Experiment with 2 to 4 clusters
    'model__linkage': ['ward', 'complete', 'average', 'single']  # Different linkage methods for clustering
}

# Default scoring metric
# Note: Silhouette score works well for evaluating convex clusters formed by hierarchical clustering.
default_scoring = 'silhouette'
````

### 2.3.2 Dimensionality Reduction
<a id="dimred"></a>

Dimensionality reduction techniques compress high-dimensional data into fewer dimensions, often for **visualization** or **speeding up** further analysis. They help reveal low-dimensional structures or simplify datasets.


#### Principal Component Analysis (PCA)
<a id="pca"></a>

**Description**:  
Principal Component Analysis finds new orthogonal axes (principal components) capturing maximal variance in the data. It’s a linear method.

**Key Characteristics**:  
- **Global Variance Preservation**: Maximizes total variance captured.  
- **Interpretability**: Each principal component is a linear combination of original features.  
- **Sensitive to Scaling**: Often requires mean-centering and scaling features.


##### Define the model (optional)

````python
%%writefile models/unsupervised/dimred/pca.py

"""
pca.py

This module defines a Principal Component Analysis (PCA) model for dimensionality reduction.
PCA is a widely used technique to reduce the dimensionality of large datasets by projecting the data
onto a lower-dimensional subspace while preserving as much variance as possible.

Key Features:
- Reduces computational complexity for high-dimensional data.
- Helps in visualizing data in 2D or 3D space.
- Useful as a preprocessing step for clustering or classification.

Parameters:
    - n_components (int, float, or None): Number of principal components to keep.
        - int: Specifies the exact number of components.
        - float: Keeps enough components to explain the specified fraction of variance (e.g., 0.95 for 95% variance).
        - None: Keeps all components (default).

Default:
    - n_components=2: Projects the data onto 2 dimensions for visualization purposes.

"""

from sklearn.decomposition import PCA

# Define the PCA estimator
estimator = PCA(n_components=2)  # Default to 2D projection for visualization
````

#### t-SNE
<a id="tsne"></a>

**Description**:  
t-Distributed Stochastic Neighbor Embedding focuses on preserving local neighborhoods, often producing visually separable 2D/3D embeddings.

**Key Characteristics**:  
- **Great for Visualization**: Commonly used for exploring high-dimensional data in 2D.  
- **Local Structure Emphasis**: Points that are close in high-dimensional space remain close in low dimension.  
- **Slow for Large Data**: Memory and time complexity can be high.


##### Define the model (optional)

````python
%%writefile models/unsupervised/dimred/tsne.py

"""
tsne.py

This module defines a t-Distributed Stochastic Neighbor Embedding (t-SNE) model
for dimensionality reduction. t-SNE is primarily used for visualizing high-dimensional
data by projecting it into a lower-dimensional space (typically 2D or 3D).

Key Features:
- Nonlinear dimensionality reduction technique.
- Preserves local relationships within the data.
- Useful for exploring clustering structures in high-dimensional datasets.

Parameters:
    - n_components (int): Number of dimensions for projection (default: 2 for visualization).
    - perplexity (float): Controls the balance between local and global data structure.
        - Typical values range between 5 and 50.
    - learning_rate (float, optional): Learning rate for optimization (default: 'auto').
    - random_state (int, optional): Ensures reproducibility of the results.

Default:
    - n_components=2: Projects the data into a 2D space for visualization purposes.
    - perplexity=30: A good starting point for most datasets.

"""

from sklearn.manifold import TSNE

# Define the t-SNE estimator
estimator = TSNE(n_components=2, perplexity=30)  # Default to 2D projection with a reasonable perplexity
````

#### UMAP
<a id="umap"></a>

**Description**:  
Uniform Manifold Approximation and Projection is a manifold-learning technique that, like t-SNE, preserves local structure but is often faster and scales better.

**Key Characteristics**:  
- **Local & Global**: Can retain both local neighborhoods and some global structure.  
- **Speed & Scalability**: Typically faster than t-SNE on larger datasets.  
- **Manifold Assumptions**: Works best when data lies on a low-dimensional manifold.


##### Define the model (optional)

### 2.3.3 Anomaly Detection
<a id="anomaly"></a>

Anomaly (or outlier) detection aims to identify rare items, events, or observations that deviate significantly from the majority. It’s widely used in **fraud detection**, **network intrusion** detection, and **manufacturing fault** detection.


#### Isolation Forest
<a id="isolation-forest"></a>

Isolation Forest isolates outliers by randomly splitting features. Points that are quicker to isolate are considered more “anomalous.”

**Key Characteristics**:  
- **Efficient**: Works well with large datasets.  
- **Random Splits**: Outliers have shorter average path lengths in the tree ensemble.  
- **Requires contamination**: Must guess fraction or use auto-estimation to define outlier proportion.


##### Define the model (optional)

````python
%%writefile models/unsupervised/anomaly/isolation_forest.py

"""
isolation_forest.py

This module defines an Isolation Forest model for anomaly detection.
Isolation Forest is an efficient and effective algorithm for identifying
outliers in high-dimensional datasets.

Key Features:
- Utilizes a tree-based approach to isolate anomalies.
- Efficient for both large datasets and high-dimensional spaces.
- Automatically determines the expected proportion of anomalies.

Parameters:
    - n_estimators (int): Number of base estimators in the ensemble.
        - Default: 100.
    - contamination (str or float): Expected proportion of outliers in the data.
        - Default: 'auto' (automatically inferred based on dataset size).
    - max_samples (int or float): Number of samples to draw for training each estimator.
        - Default: 'auto' (uses min(256, number of samples)).

Default Configuration:
    - n_estimators=100: Adequate for most datasets.
    - contamination='auto': Automatically estimates the proportion of outliers.
"""

from sklearn.ensemble import IsolationForest

# Define the Isolation Forest estimator
estimator = IsolationForest(
    n_estimators=100,       # Default number of trees
    contamination='auto',   # Automatically estimates the contamination proportion
    random_state=42         # Ensures reproducibility
)
````

#### One-Class SVM
<a id="one-class-svm"></a>

**Description**:  
One-Class SVM learns a decision boundary around the majority of data, labeling points outside that boundary as outliers.

**Key Characteristics**:  
- **Kernel Flexibility**: Adapts to non-linear shapes.  
- **Nu Parameter**: Controls outlier fraction / boundary tightness.  
- **Potentially High Complexity**: Larger memory/time cost for big datasets.


##### Define the model (optional)

````python
%%writefile models/unsupervised/anomaly/one_class_svm.py

"""
one_class_svm.py

This module defines a One-Class SVM model for anomaly detection.
One-Class SVM identifies a decision boundary that separates normal data points from potential outliers
in a high-dimensional feature space.

Key Features:
- Effective for detecting anomalies in high-dimensional datasets.
- Flexible kernel options for nonlinear decision boundaries.
- Suitable for datasets with a small proportion of outliers.

Parameters:
    - kernel (str): Specifies the kernel type used in the algorithm.
        - Common options: 'linear', 'poly', 'rbf' (default), and 'sigmoid'.
    - gamma (str or float): Kernel coefficient. Determines the influence of each sample.
        - Default: 'scale' (1 / (n_features * X.var())).
    - nu (float): Approximate fraction of outliers in the dataset.
        - Must be in the range (0, 1]. Default: 0.05 (5% of data considered outliers).

Default Configuration:
    - kernel='rbf': Radial Basis Function for nonlinear separation.
    - gamma='scale': Automatically adjusts kernel influence based on dataset features.
    - nu=0.05: Assumes approximately 5% of data points are outliers.
"""

from sklearn.svm import OneClassSVM

# Define the One-Class SVM estimator
estimator = OneClassSVM(
    kernel='rbf',  # Radial Basis Function kernel for nonlinear boundaries
    gamma='scale',  # Adjusts kernel influence based on dataset variance
    nu=0.05         # Assumes 5% of the data are outliers
)
````

#### Local Outlier Factor (LOF)
<a id="local-outlier-factor"></a>

**Description**:  
LOF scores each point based on its local density compared to the densities of its neighbors. Lower local density indicates higher outlierness.

**Key Characteristics**:  
- **No Fit/Predict**: Typically done with `.fit_predict(X)` on the same data (unless `novelty=True`).  
- **Neighbor Density**: Relies on local comparisons.  
- **Not for Novelty**: Classic LOF can’t easily be applied to new data unless `novelty=True`.


##### Define the model (optional)

````python
%%writefile models/unsupervised/anomaly/local_outlier_factor.py

"""
local_outlier_factor.py

This module defines a Local Outlier Factor (LOF) model for anomaly detection.
LOF identifies anomalies by comparing the local density of a sample to the density
of its neighbors. Samples with significantly lower density are flagged as outliers.

Key Features:
- Detects local anomalies in datasets with varying densities.
- Effective for datasets where the notion of an outlier is context-dependent.
- Non-parametric method that adapts to the data's structure.

Parameters:
    - n_neighbors (int): Number of neighbors used to calculate local density.
        - Default: 20. Higher values smooth out anomalies but may miss local patterns.
    - contamination (str or float): Proportion of outliers in the data.
        - 'auto': Automatically estimates the proportion based on the dataset size.
        - float: Manually set the expected proportion (e.g., 0.1 for 10%).
    - novelty (bool): If True, allows the model to be applied to new unseen data.

Limitations:
- LOF directly computes predictions during `fit_predict()` and does not support `predict()`
  unless `novelty=True`.

Default Configuration:
    - n_neighbors=20: Uses 20 neighbors for density comparison.
    - contamination='auto': Automatically estimates the proportion of outliers.
    - novelty=True: Enables predictions on unseen data.
"""

from sklearn.neighbors import LocalOutlierFactor

# Define the Local Outlier Factor estimator
estimator = LocalOutlierFactor(
    n_neighbors=20,  # Number of neighbors to calculate density
    contamination='auto',  # Auto-detect the proportion of outliers
    novelty=True  # Enables prediction on new data
)
````

### Step 3: Clustering Example (e.g., K-Means)

We'll illustrate a simple approach:
1. Create a pipeline for optional preprocessing (imputation/scaling).
2. Define a clustering estimator (K-Means).
3. (Optionally) do silhouette-based hyperparameter tuning for `n_clusters`.
4. Fit the final model and retrieve cluster labels.
5. (Optionally) evaluate silhouette score and visualize if 2D or 3D.

Feel free to expand or modify the parameter grid for more advanced logic.

````python
from sklearn.cluster import KMeans
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import silhouette_score
from sklearn.base import clone

# Constants for hyperparameter tuning
CV_FOLD = 5  # Placeholder for potential cross-validation use (not implemented here)
best_score = -1  # Initialize the best silhouette score to a low value
best_params = None  # Placeholder to store the best hyperparameters
best_estimator = None  # Placeholder to store the best model (fitted estimator)

# Initialize the KMeans clustering estimator
estimator = KMeans(
    n_clusters=3,        # Default number of clusters (will be tuned)
    random_state=42      # Ensures reproducibility of results
)

# Define the hyperparameter grid for KMeans tuning
param_grid = {
    'model__n_clusters': [2, 3, 4, 5],  # Test different numbers of clusters (2 to 5)
    'model__init': ['k-means++', 'random'],  # Compare optimized ('k-means++') vs random initialization
    'model__n_init': [10, 20, 50]  # Vary the number of initializations to improve clustering stability
}

# Hyperparameters:
# - 'model__n_clusters': The number of clusters to form and the number of centroids to generate.
#   Larger values can capture more complex data structures but may lead to overfitting.
# - 'model__init': 'k-means++' selects initial cluster centers to speed up convergence, while 'random'
#   chooses centers randomly, which may require more iterations to converge.
# - 'model__n_init': Specifies how many times the k-means algorithm will run with different
#   centroid seeds. Higher values can provide better stability but increase computation time.
````

````python
print("Performing hyperparameter tuning...")

# Start timer to measure the duration of the hyperparameter tuning process
start_time = timer()

# Iterate through all combinations of hyperparameters in the parameter grid
for params in ParameterGrid(param_grid):
    # Clone the base estimator to ensure a fresh model for each parameter combination
    current_estimator = clone(estimator)

    # Apply the current set of parameters to the cloned estimator
    for param, val in params.items():
        path = param.split('__')  # Split the parameter name (e.g., 'model__n_clusters')
        if len(path) > 1:
            # If parameter name contains '__', it's a nested parameter (e.g., model attributes)
            setattr(current_estimator, path[1], val)
        else:
            # Directly set parameter on the estimator (non-nested)
            setattr(current_estimator, param, val)

    # Fit the current estimator on the dataset
    current_estimator.fit(X)

    # Retrieve cluster labels for evaluation
    if hasattr(current_estimator, 'labels_') and current_estimator.labels_ is not None:
        labels = current_estimator.labels_  # Use `labels_` attribute if available
    elif hasattr(current_estimator, 'predict'):
        labels = current_estimator.predict(X)  # Otherwise, use the `predict()` method
    else:
        # Raise an error if the estimator does not provide a valid way to retrieve labels
        raise ValueError("No valid way to retrieve cluster labels for this estimator.")

    # Evaluate clustering performance using the silhouette score
    unique_labels = set(labels)  # Check the number of unique clusters
    if len(unique_labels) > 1:
        # Compute the silhouette score if more than one cluster is found
        score = silhouette_score(X, labels)
    else:
        # Assign a low score if only one cluster is found (invalid clustering)
        score = -999

    # Update the best parameters and estimator if the current score is higher
    if score > best_score:
        best_score = score
        best_params = params
        best_estimator = current_estimator

# Handle the case where no parameter combination produces meaningful clusters
if best_estimator is None:
    print("No valid parameter combination produced more than 1 cluster. Using the original estimator.")
    estimator.fit(X)  # Fit the original estimator with default parameters
    best_params = {}
else:
    # Print the best silhouette score and parameters
    print(f"Best silhouette score: {best_score:.4f}")
    estimator = best_estimator  # Update the estimator to the best fitted model

# End timer and calculate the duration of hyperparameter tuning
end_time = timer()
train_time = end_time - start_time
print(f"Hyperparameter tuning completed in {train_time:.2f} seconds.")
print(f"Best Parameters: {best_params}")
````

````python
# Define the directory for saving clustering results
# Create a specific subdirectory within the general results path for clustering-related files
results_clustering_path = os.path.join(RESULTS_PATH, "clustering")

# Ensure the directory exists (create it if it doesn't already)
os.makedirs(results_clustering_path, exist_ok=True)

# Evaluate the clustering results using the best model
# Determine cluster labels from the estimator
if hasattr(estimator, 'labels_'):
    # Use `labels_` if available (e.g., KMeans)
    labels = estimator.labels_
else:
    # Use `predict()` for models that do not provide `labels_` (e.g., GaussianMixture)
    labels = estimator.predict(X)

# Analyze the uniqueness of cluster labels
unique_labels = set(labels)  # Extract unique cluster labels to check for meaningful clustering
if len(unique_labels) > 1:
    # Compute the silhouette score only if more than one cluster is present
    sil = silhouette_score(X, labels)
    print(f"Final Silhouette Score: {sil:.4f}")

    # Save the silhouette score to a CSV file
    metrics_path = os.path.join(results_clustering_path, "metrics.csv")
    pd.DataFrame({"Silhouette": [sil]}).to_csv(metrics_path, index=False)
    print(f"Metrics saved to {metrics_path}")
else:
    # Handle cases where only one cluster is formed (invalid clustering scenario)
    print("Only one cluster found; silhouette score not meaningful.")
````

````python
# Visualization of Clusters
print("Creating cluster visualization...")

# Handle cases with low-dimensional or no features
if X.shape[1] < 2:
    # Check if data is 1-dimensional
    if X.shape[1] == 1:
        print("Only 1 feature available; cannot create a 2D scatter plot.")
    else:
        # No features available for visualization
        print("No features available for plotting.")
else:
    # For high-dimensional data (>2 features), reduce dimensions to 2D using PCA
    if X.shape[1] > 2:
        from sklearn.decomposition import PCA
        pca = PCA(n_components=2)  # Reduce data to 2 components for visualization
        X_2d = pca.fit_transform(X)

        # Calculate and annotate variance explained by each principal component
        var_ratio = pca.explained_variance_ratio_  # Variance ratios for PCs
        pc1_var = var_ratio[0] * 100  # Variance explained by PC1 in percentage
        pc2_var = var_ratio[1] * 100  # Variance explained by PC2 in percentage
        x_label = f"PC1 ({pc1_var:.2f}% variance)"
        y_label = f"PC2 ({pc2_var:.2f}% variance)"
    else:
        # For 2D data, use feature names if available for axis labels
        if 'df' in locals() and df.shape[1] == 2:  # Check if DataFrame `df` exists
            x_label = df.columns[0]  # Name of the first feature
            y_label = df.columns[1]  # Name of the second feature
        else:
            # Generic labels if feature names are unavailable
            x_label = "Feature 1"
            y_label = "Feature 2"
        X_2d = X  # Use original 2D data directly

# Create a scatter plot of the clusters
plt.figure(figsize=(6, 5))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='viridis', s=30)  # Color points by cluster labels
plt.title(f"{estimator.__class__.__name__} Clusters")
plt.xlabel(x_label)
plt.ylabel(y_label)

# Save the cluster visualization plot
plot_path = os.path.join(results_clustering_path, "clusters.png")
plt.savefig(plot_path)  # Save the plot as a PNG file
plt.show()  # Display the plot
print(f"Cluster plot saved to {plot_path}")
````

````python
from sklearn.decomposition import PCA

# Step 1: Initialize the PCA estimator
# PCA is used to reduce the dimensionality of the data for visualization or preprocessing
# Here, we reduce the data to 2 dimensions (2D) for visualization purposes
estimator = PCA(n_components=2)

# Step 2: Apply PCA to the data
# Fit the PCA model to the data and transform it into a lower-dimensional space
start_time = timer()  # Record the start time for performance measurement
X_transformed = estimator.fit_transform(X)  # Perform PCA transformation
end_time = timer()  # Record the end time

# Calculate the time taken for dimensionality reduction
train_time = end_time - start_time

# Step 3: Display results
# Print the time taken for PCA and the change in dimensionality
print(f"Dimensionality Reduction with {estimator.__class__.__name__} completed in {train_time:.2f} seconds.")
print("Original data shape:", X.shape, "→ Transformed data shape:", X_transformed.shape)

# Step 4: Define the directory for saving results
# Create a subdirectory for dimensionality reduction results within the results path
results_dimred_path = os.path.join(RESULTS_PATH, "DimRed")

# Ensure the directory exists (create it if it doesn't)
os.makedirs(results_dimred_path, exist_ok=True)
````

### Step 5: Anomaly Detection Example (e.g., Isolation Forest)

Finally, we demonstrate anomaly detection:
1. Create pipeline if needed (imputation/scaling).
2. Fit the anomaly detector (Isolation Forest).
3. Convert raw predictions (+1 inlier, -1 outlier) into (0,1).
4. Save predictions and model.

We'll skip param grids here, but you can adapt a manual approach if you want.

````python
from sklearn.ensemble import IsolationForest

# Initialize the IsolationForest model
estimator = IsolationForest(n_estimators=100, random_state=42)

# Step 1: Train the Isolation Forest model
start_time = timer()  # Record the start time
estimator.fit(X)  # Fit the model to the dataset
end_time = timer()  # Record the end time
train_time = end_time - start_time  # Calculate training time
print(f"Anomaly detection training with {estimator.__class__.__name__} completed in {train_time:.2f} seconds.")

# Step 2: Predict anomalies
# The predict method typically returns:
#   +1 for normal (inlier) samples
#   -1 for anomalies (outlier) samples
raw_preds = estimator.predict(X)

# Step 3: Convert raw predictions to binary format:
#   0 = normal (inlier), 1 = anomaly (outlier)
preds_binary = np.where(raw_preds == 1, 0, 1)  # Convert +1 to 0, and -1 to 1

# Step 4: Count outliers and inliers
outlier_count = np.sum(preds_binary)  # Count the number of anomalies
inlier_count = len(preds_binary) - outlier_count  # Calculate the number of normal samples
print(f"Detected {outlier_count} outliers out of {len(X)} samples ({inlier_count} normal).")

# Step 5: Define the directory to save anomaly detection results
results_anomaly_path = os.path.join(RESULTS_PATH, "anomaly")
# Create the directory if it doesn't already exist
os.makedirs(results_anomaly_path, exist_ok=True)
````

````python
# Visualization for Anomaly Detection (2D or 3D)
print("Creating anomaly detection visualization...")

# Check the dimensionality of the dataset
if X.shape[1] < 2:
    # If the dataset has less than 2 features, skip visualization
    print("Dataset has less than 2 features; 2D scatter plot not possible. Skipping visualization.")
else:
    # Reduce dimensions to 2D if the dataset has more than 2 features
    if X.shape[1] > 2:
        from sklearn.decomposition import PCA
        pca = PCA(n_components=2)  # Reduce to 2 principal components
        X_2d = pca.fit_transform(X)  # Apply PCA transformation
        x_label = "Principal Component 1"  # Label for x-axis
        y_label = "Principal Component 2"  # Label for y-axis
    else:
        # If the dataset already has 2 features, use it directly
        X_2d = X
        x_label = df.columns[0] if df.shape[1] >= 1 else "Feature 1"  # Default or use column name
        y_label = df.columns[1] if df.shape[1] >= 2 else "Feature 2"  # Default or use column name

    # Create scatter plot
    plt.figure(figsize=(6, 5))
    # Color points differently for outliers (red) and normal points (blue)
    colors = np.where(preds_binary == 1, 'r', 'b')  # Red for outliers, blue for normal
    plt.scatter(X_2d[:, 0], X_2d[:, 1], c=colors, s=30, alpha=0.7)
    plt.title(f"{estimator.__class__.__name__} Anomaly Detection")
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    # Save the plot
    plot_path = os.path.join(results_anomaly_path, "anomaly_plot.png")
    os.makedirs(results_anomaly_path, exist_ok=True)  # Ensure the directory exists
    plt.savefig(plot_path)
    plt.show()
    print(f"Anomaly plot saved to {plot_path}")
````
